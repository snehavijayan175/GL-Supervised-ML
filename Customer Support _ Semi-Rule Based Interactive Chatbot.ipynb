**Extracting data from json data file**

```python
words = []
labels = []
docs_x = []
docs_y = []

for intent in data['intents']:
  for pattern in intent['patterns']:
    wrds = nltk.word_tokenize(pattern)
    words.extend(wrds)
    docs_x.append(wrds)
    docs_y.append(intent["tag"])
        
  if intent['tag'] not in labels:
    labels.append(intent['tag'])
```

```python
labels
```

**Performing Stemming on the words**

```python
words = [stemmer.stem(w.lower()) for w in words if w != "?"]
words = sorted(list(set(words)))

labels = sorted(labels)
```

**Creating Bag of words for vector representation**

```python
training = []
output = []

out_empty = [0 for _ in range(len(labels))]

for x, doc in enumerate(docs_x):
  bag = []
  wrds = [stemmer.stem(w.lower()) for w in doc]

  for w in words:
    if w in wrds:
      bag.append(1)
    else:
      bag.append(0)

  output_row = out_empty[:]
  output_row[labels.index(docs_y[x])] = 1

  training.append(bag)
  output.append(output_row)
```

```python
training = numpy.array(training)
output = numpy.array(output)
```

**Neural Network**

```python
model=Sequential()
model.add(Dense(100,input_dim=len(training[0]),activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10,activation='softmax'))
```

**Compiling and training the model**

```python
sgd = optimizers.SGD(lr=0.001, decay=1e-6, momentum=0.9)
adam = tensorflow.keras.optimizers.Adam(learning_rate=0.1)
model.compile(optimizer=adam, loss='categorical_crossentropy', metrics=["accuracy"])
```

```python
history = model.fit(training, output, epochs = 256) #batch_size = 32
```

**Predicting from given model by first converting it into bag of words**

```python
def bag_of_words(s, words):
  bag = [0 for _ in range(len(words))]
  #bag=[]

  s_words = nltk.word_tokenize(s)
  s_words = [stemmer.stem(word.lower()) for word in s_words]

  for se in s_words:
    for i, w in enumerate(words):
      if w == se:
        bag[i] = 1
      
  return np.array(bag)
```
